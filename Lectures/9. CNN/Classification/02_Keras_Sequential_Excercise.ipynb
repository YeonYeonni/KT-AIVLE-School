```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
```

## Linear Regression

```python
from sklearn.datasets import load_boston
```

```python
# Data Load
boston = load_boston()
```

```python
x = boston.data
y = boston.target

x.shape, y.shape
```

```python
# 1. Session Clear
keras.backend.clear_session()

# 2. Model
model_reg = keras.models.Sequential()

# 3. Model Block
model_reg.add(keras.layers.Input(shape=(13,)))
model_reg.add(keras.layers.Dense(1))

# 4. Compile
model_reg.compile(loss='mse', optimizer='adam')
```

```python
# 5. fit
model_reg.fit(x, y, epochs=10, verbose=1)
```

```python
# 6. predict
print(y[:10])
print(model_reg.predict(x)[:10].reshape(-1))
```

## Logistic Regression

```python
from sklearn.datasets import load_breast_cancer
```

```python
# Data Load
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target

x.shape, y.shape
```

```python
# 1. Session Clear
keras.backend.clear_session()

# 2. Model
model_class = keras.models.Sequential()

# 3. Model Block
model_class.add(keras.layers.Input(shape=(30,)))
model_class.add(keras.layers.Dense(1, activation='sigmoid'))

# 4. Compile
model_class.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 5. fit
model_class.fit(x, y, epochs=10, verbose=1)
```

```python
# 6. predict

print(y[:10])
print(model_class.predict(x)[:10].reshape(-1))
```
