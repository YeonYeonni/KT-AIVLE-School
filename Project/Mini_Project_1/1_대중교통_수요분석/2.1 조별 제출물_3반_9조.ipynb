* 가설1. 주말심야이동평균인구와 정류장 수간에 상관관계가 있다 (도착지 기준)

```python
data_1 = pd.merge(data1, data3, on=['자치구', '요일'], how='inner')
```

```python
data_1.head(10)
```

```python
# 주말 필터링
data_1weekend = data_1.loc[(data_1['요일'] == '토') | (data_1['요일'] == '일')]
```

```python
data_1weekend.corr() #상관계수 확인
```

```python
spst.pearsonr(data_1weekend['정류장수'],data_1weekend['심야이동인구평균'])
```

```python
spst.pearsonr(data_1weekend['정류장수'],data_1weekend['심야이동인구합'])
```

```python
a = data_1weekend.groupby('자치구', as_index=False)['정류장수'].sum()
b = data_1weekend.groupby('자치구', as_index=False)['심야이동인구합'].sum()
```

```python
data_1with2_weekend_res = pd.merge(a, b, on='자치구', how='inner')
data_1with2_weekend_res['심야이동인구합 대비 정류장수'] = data_1with2_weekend_res['정류장수'] / data_1with2_weekend_res['심야이동인구합']
data_1with2_weekend_res = data_1with2_weekend_res.sort_values('심야이동인구합 대비 정류장수',ascending=True)
data_1with2_weekend_res
```

#### 결론) 심야이동인구합 대비 정류장 수가 적을수록 정류장 수를 늘려야 한다
